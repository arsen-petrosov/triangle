```python
a = float(input())
b = float(input())
c = float(input())

if (a+b>c and a+c>b and b+c>a):
   
    maxSide = max(a, b, c)
    if maxSide == a:
        cos = (b**2+c**2-a**2)/(2*b*c)
    elif maxSide == c:
        cos = (b**2+a**2-c**2)/(2*b*a)
    else:
        cos = (a**2+c**2-b**2)/(2*a*c)

        
    if ((a == b) or (a == c) or (b == c)):
        if a == b and a == c:
            print ("равносторонний")
        else:
            if cos < 0:
                print ("равнобедренный, тупоугольный")
            elif cos == 0:
                print ("равнобедренный, прямоугольный")
            else:
                print ("равнобедренный, остроугольный")
            
            
    else:
        if cos < 0:
            print ("тупоугольный")
        elif cos == 0:
            print ("прямоугольный")
        else:
            print ("остроугольный")
else:
    print ("такого треугольника не существует")
```
